##DataFrames

- A Series is a single column in a DataFrame, and can be created with a list
  - However, a Series is not a list! It has an index and datatype (int, float, object)
- Lists in python are enclosed in [ ] and can hold many elements within a single variable

Let's create a series using the series function from python! As you will see, dataframes are indexed starting at 0

```python
# first_list and second_list are both Lists, not Series
first_list = ["a", "b", "c", "d", "e"]
second_list = [1, 2, 3, 4]
# use pd.Series(data) below and then run this code cell! 
# Note that in the call, data = list you want to use
pd.Series(second_list)
```


- When creating a series, **you can also use a dictionary key/value object**
- This is the convensional way to create a dataframe when you are entering your own elements

**A dataframe essetially is a group of Series put together**

```python
rating_list = [9, 0, 7, 8]
company = ["It's happening", "Riot Games", "Behaviour Interactive", "Mojang Studios"]
# Combine the two into a dataframe below using the pandas DataFrame() method
# pd.DataFrame({"Column 1": data1, "Column 2": data2})
pd.DataFrame({"Rating": rating_list, "Company": company})
```

Let's assign a value to the **index** of the previous dataframe we created
- We can do this by using an index parameter in the DataFrame constructor

```python
# pd.DataFrame({"Column 1": data1, "Column 2": data2}, index = new_index)
game_list = ["PlateUp!", "League Of Legends", "Dead By Daylight", "Minecraft"]
pd.DataFrame({"Rating": rating_list, "Company": company}, index=game_list)
```

# Filtering Our Data using LOC



We can select data in our dataframe by choosing a specific value we want to view

A dataframe can be queried:

- by label: loc (location)
- by index position: iloc (using slices / integer location-- we will not get into this one today :) )

Arguments to loc can be:
- single values
- slices
- lists

The first argument corresponds to rows, the second argument 
corresponds to columns.

```python
# Lets say we only want to see albums released in the year 2000: Use column named 'Release Year'
# Compare functions such as == > < can be used on columns 
# df.loc[df['column_name'] == xxxx, ['column_name1', 'column_name2']]

df.loc[df['Release Year'] == 2000, ['Artist', 'Title']]
```

```python
# Let's find data that are released in the year 2000 and in the month of September
# We can enclose multiple statements in parenthesis and connect them using '&' as AND, we can also use '|' for OR

# df.loc[(df['column_name'] == 'this_string') & (df['column_name2'] == 'this_string2')]
df.loc[(df['Release Year'] == 2000) & (df['Release Month'] == 'September')]
```


Using str.contains(), we can access the string inside of the column to check its contents

Let's find album **titles** with the string 'Egg' in them 

```python
# df.loc[df["column_name"].str.contains('string')]
df.loc[df["Title"].str.contains('Egg')]
```

Filtering to only look at albums released after 2009

```python
# Compare functions such as == > < can be used on columns 
# df.loc[df['column_name'] > x]

df.loc[df['Release Year'] > 2009]
```

Now lets find Albums within the Pop Rock or Rock **genre** that have a **metacritic score** of over 60/100

```python
# We can enclose multiple statements in parenthesis and connect them using '&' as AND, we can also use '|' for OR
# df.loc[(df['column_name'] == 'this_string') & (df['column_name2'] == 'this_string2')]

df.loc[((df["Genre"] == 'Pop Rock') | (df["Genre"] == 'Rock')) & (df["Metacritic Critic Score"] > 60)]
```

Find information of the artist with the highestmetacritic score 

```python
# To find the maximum value of a column and to return its corresponding row values
# we can use df.loc[df['col_name'].idxmax()]
df.loc[df['Metacritic Critic Score'].idxmax()]
```

# Relationship Plots

Relationship plots can be used to show the correlation between 2 columns, denoted by x and y - the axis of the plots

```python
# Let's graph the relationship between release year and meta critic score of Grass type pokemon
# First let's create a subset of the dataframe for just grass types (For Type 1)
grass_types = pokemon[pokemon['Type 1'] == "Grass"]
grass_types
```

```python
# Next let's create the graph of the relationship between Defense and Attack
# RelPlots can be created by executing the following 2 commands: 
# sns.relplot(data=df, x="x_value", y="y_value")
# plt.title("Title");

sns.relplot(data=grass_types,  x='Defense', y='Attack')
plt.title("Grass Type Defense and Attack")
```

By setting the hue argument in relplot(), we can specify which column we want to determine the color of each point

```python
# Let's add a hue argument to represent the pokemon's HP
sns.relplot(data=grass_types,  
            x='Defense', 
            y='Attack',
            hue = 'HP')
plt.title("Grass Type Stats")
```

**We can change the color palette of your graphs by setting the palette argument in relplot**

Palette options can be found here: [seaborn_color_palettes](http://seaborn.pydata.org/tutorial/color_palettes.html)

*Some options from link above: flare, crest, magma, viridis, rocket, Spectral, coolwarm, Purples*

You can also reverse the palette by adding "_r" to the end of the color name 

EX: palette = "crest_r"

```python
# Let's add a palette to colorize the HP to something else :)
sns.relplot(data=grass_types,  
            x='Defense', 
            y='Attack',
            hue='HP',
            palette='Greens')
plt.title("Grass Type Stats")
```

We can also introduce the size argument, which like hue can be used to express another data 

Let's use size to show the **Speed Stats**

```python
# To do this, introduce the following arguments to the relplot call 
# size = "column name",
# sizes = (min_size, max_size)

sns.relplot(data=grass_types,  
            x='Defense', 
            y='Attack',
            hue='HP',
            palette='Greens',
            size='Speed',
            sizes=(5,200)
            )
plt.title("Grass Type Stats")
```
